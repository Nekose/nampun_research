```python
import numpy as np
import pandas as pd
import matplotlib
import math
import skbio

indels="df_indels_combined_32720.csv"
rearange="rearrangement_updated_32720.csv"
```

```python
df=pd.read_csv(rearange)
df2=pd.read_csv(indels)
```

```python
display(df)
display(df2)
```

```python
#Renaming all of the columns in DF 1 to match the conventions in the new data (NHLC_1A --> 1A)
df.rename(columns=lambda x: x.lstrip("NHLC_"), inplace=True)
display(df)
```

```python
#print(df.loc[df['type'] == 'wt'])
```

```python
for i in range(len(df2)):
    eventtype = df2.iloc[i]["label"]
    samplename = df2.iloc[i]["name"]
    readcount = df2.iloc[i]["read_count"]
    print(eventtype)
    #print(df['type'])
    if (df['type'] == eventtype).any():
        print("found it")
        df.loc[df['type'] == eventtype , [samplename]] += readcount
    else:
        print("New Type")
        df = df.append({'type':eventtype, samplename: readcount},ignore_index=True)
print(len(df))
```

```python
df.fillna(0, inplace=True)
display(df)
```

```python
f = open("indel_rearange_combined.csv", "a")
f.write(df.to_csv(index=False))
f.close()
```
